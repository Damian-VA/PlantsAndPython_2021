```python
# Belowground vs Aboveground 
# using the above/below.csv (https://docs.google.com/spreadsheets/d/1kive2B1lAYwv3SS2HKyB5dz5jad_0IZw9tDYWHaw1j4/edit?usp=sharing)

# Import the dataset as above_below
above_below = pd.read_csv("./result_table_100_renamed_tissues.csv", encoding= 'unicode_escape')

# First check how many tissue types you begin with (should be 104 for the 100 genes dataset)
print("Before replacing with above/belowground labels, in the 100 genes dataset there're", len(above_below['Tissue'].unique()), "tissue types")

# Replace all belowground tissue levels with the 'belowground' (use the above/below.csv as basis)
above_below['Tissue']= above_below['Tissue'].replace({
    "root":"belowground",
    "hypocotyl":"belowground",
    "root tip":"belowground",
    "Root":"belowground",
    "roots":"belowground",
    "root protoplasts":"belowground",
    "root QC cells":"belowground",
    "Hypocotyls":"belowground",
    "whole root":"belowground",
    "7-day-old root tips":"belowground",
    "1cm long root tips":"belowground",
    "root tips":"belowground",
    "root hair":"belowground",
    "primary root":"belowground",
    "Apical portions of plants (cut in hypocotyl)":"belowground",
    "whole roots":"belowground",
    "root tissue":"belowground",
    "roots and hyphae":"belowground",
    "Roots":"belowground",
    "root apex":"belowground"})

# Now check how many tissue types there're (should be 98 levels)
print("After replacing all belowground labels, there're", len(above_below['Tissue'].unique()), "tissue types")

# Replace all the non-belowground tissue levels with 'aboveground'
above_below.loc[above_below['Tissue'] != 'belowground', 'Tissue'] = 'aboveground'

# Now check how many tissue types there're (should be 2 levels)
print("After replacing all below/aboveground labels, there're", len(above_below['Tissue'].unique()), "tissue types")

# Check how many samples (rows) per tissue category
print(above_below['Tissue'].value_counts())

# Save the new dataset
above_below.to_csv('below_above.csv')
```

```python
# Belowground, aboveground reproductive, aboveground vegetative
# using the above/below.csv (https://docs.google.com/spreadsheets/d/1kive2B1lAYwv3SS2HKyB5dz5jad_0IZw9tDYWHaw1j4/edit?usp=sharing)

# Import the dataset as above_below
above_below = pd.read_csv("./result_table_100_renamed_tissues.csv", encoding= 'unicode_escape')

# First check how many tissue types you begin with (should be 104 for the 100 genes dataset)
print("Before replacing with above/belowground labels, in the 100 genes dataset there're", len(above_below['Tissue'].unique()), "tissue types")

# Replace all belowground tissue levels with the 'belowground' (use the above/below.csv as basis)
above_below['Tissue']= above_below['Tissue'].replace({
    "root":"belowground",
    "hypocotyl":"belowground",
    "root tip":"belowground",
    "Root":"belowground",
    "roots":"belowground",
    "root protoplasts":"belowground",
    "root QC cells":"belowground",
    "Hypocotyls":"belowground",
    "whole root":"belowground",
    "7-day-old root tips":"belowground",
    "1cm long root tips":"belowground",
    "root tips":"belowground",
    "root hair":"belowground",
    "primary root":"belowground",
    "Apical portions of plants (cut in hypocotyl)":"belowground",
    "whole roots":"belowground",
    "root tissue":"belowground",
    "roots and hyphae":"belowground",
    "Roots":"belowground",
    "root apex":"belowground"})

# Now check how many tissue types there're (should be 98 levels)
print("After replacing belowground there're", len(above_below['Tissue'].unique()), "tissue types")

# Replace all aboveground reproductive tissue levels with 'aboveground reproductive'
above_below['Tissue']= above_below['Tissue'].replace({
"inflorescence":"aboveground reproductive",
"inflorescences":"aboveground reproductive",
"Inflorescence axis":"aboveground reproductive",
"Inflorescence stem":"aboveground reproductive",
"inflorescence (stage 1-12) immature flower buds":"aboveground reproductive",
"Inflorescence meristem":"aboveground reproductive",
"flower buds":"aboveground reproductive",
"flower bud":"aboveground reproductive",
"flower pedicel":"aboveground reproductive",
"unopened flower buds":"aboveground reproductive",
"flower receptacles":"aboveground reproductive",
"inflorescences composed of all unopened floral buds":"aboveground reproductive",
"flower":"aboveground reproductive",
"Flower":"aboveground reproductive",
"stigma":"aboveground reproductive",
"flower meristems":"aboveground reproductive",
"anther":"aboveground reproductive",
"carpel":"aboveground reproductive",
"carpels":"aboveground reproductive",
"stamens":"aboveground reproductive",
"stamen":"aboveground reproductive",
"stamen filament":"aboveground reproductive",
"meiocytes":"aboveground reproductive",
"gynoecia":"aboveground reproductive",
"sepal":"aboveground reproductive",
"petal":"aboveground reproductive",
"pistils":"aboveground reproductive",
"Pollinated stigma":"aboveground reproductive"})

# Now check how many tissue types there're (should be 82 levels)
print("After replacing below/aboveground reproductive there're", len(above_below['Tissue'].unique()), "tissue types")

# Replace all aboveground vegetative tissue levels (all the remaining levels) with 'abovground vegetative'
above_below.loc[~above_below['Tissue'].isin(['belowground','aboveground reproductive']), 'Tissue'] = 'aboveground vegetative'

# Now check how many tissue types there're (should be 3 levels)
print("After replacing below/aboveground reproductive/vegetative there're", len(above_below['Tissue'].unique()), "tissue types")

# Check how many samples (rows) per tissue category
print(above_below['Tissue'].value_counts())

# Save the new dataset
above_below.to_csv('below_repr_veg.csv')
```
